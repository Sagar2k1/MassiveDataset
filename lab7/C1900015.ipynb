# 1. Data

```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import functions as func
```

```python
sc = SparkContext("local", "DataFrame")
```

```python
sqlc = SQLContext(sc)
sqlc
```

```python
df = sqlc.read.csv(path, header = True)
```

```python
df.show()
```

```python
col_name = df.columns
print(col_name)
```

```python
label_col = df[0]
features_col = [df[i] for i in range(1, len(col_name))]
```

```python
names = ['label', 'features']
```

```python
from pyspark.sql.functions import col, lit, monotonically_increasing_id
from pyspark.ml.linalg import Vectors
```

```python
data = df.rdd.flatMap(lambda x: [(int(x[0]),Vectors.dense([int(x[i]) for i in range(1,len(col_name))]))]).collect()
```

```python
df2 = sqlc.createDataFrame(data,names)
df2
```

```python
df2.printSchema()
```

```python
df2.head(20)
```

```python
df3 = df2.withColumn('idx', monotonically_increasing_id())
df3.show(20)
```

# 2. K-Mean

```python
from pyspark.ml import clustering
```

```python
k = 10
maxIter = 10
Kmean = clustering.KMeans(k = k, maxIter = maxIter)
```

```python
Kmean.setWeightCol("label")
Kmean.fit(df3)
```

```python
Kmean.setPredictionCol("Prediction")
Kmean.getPredictionCol()
```

```python
transformed = Kmean.transform(df3).select("features", "Prediction")
```
